## 1. Backup Current Project

Create a comprehensive backup of the current project, including all files, dependencies, and configurations. This ensures the existing Streamlit version is preserved and can be restored if needed.

### Tasks:
- Archive the entire current project directory
- Document all current dependencies and versions
- Create a snapshot of the database (if applicable)
- Store backup in a safe, accessible location

```python
import os
import shutil
from datetime import datetime
import json

# Define project paths
current_project_path = "/Users/gangumallarajanreddy/Downloads/Student_Clustering_Project(US)"
backup_dir = os.path.join(current_project_path, "backups")
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
backup_name = f"project_backup_{timestamp}"
backup_path = os.path.join(backup_dir, backup_name)

# Create backup directory if it doesn't exist
os.makedirs(backup_dir, exist_ok=True)

print(f"Backup path will be: {backup_path}")
print(f"Timestamp: {timestamp}")
```

```python
# Function to create backup
def create_project_backup(source_path, backup_path, exclude_dirs=None):
    """
    Create a backup of the project directory
    
    Parameters:
    - source_path: Path to the project to backup
    - backup_path: Path where backup will be stored
    - exclude_dirs: List of directories to exclude from backup
    """
    if exclude_dirs is None:
        exclude_dirs = ['__pycache__', '.git', '.pytest_cache', 'backups', '.venv', 'venv']
    
    def ignore_patterns(directory, contents):
        ignored = []
        for item in contents:
            if item in exclude_dirs or item.startswith('.'):
                ignored.append(item)
        return set(ignored)
    
    try:
        shutil.copytree(source_path, backup_path, ignore=ignore_patterns)
        print(f"✓ Backup created successfully at: {backup_path}")
        return True
    except Exception as e:
        print(f"✗ Error creating backup: {str(e)}")
        return False

# Execute backup
backup_success = create_project_backup(current_project_path, backup_path)
```

```python
# Document current dependencies
import subprocess

requirements_path = os.path.join(current_project_path, "requirements.txt")
dependencies_log_path = os.path.join(backup_dir, f"dependencies_{timestamp}.txt")

def document_dependencies():
    """Capture and log current project dependencies"""
    try:
        # Try to read existing requirements.txt
        if os.path.exists(requirements_path):
            with open(requirements_path, 'r') as f:
                current_deps = f.read()
            
            # Save to backup log
            with open(dependencies_log_path, 'w') as f:
                f.write(f"Dependencies Backup - {timestamp}\n")
                f.write("=" * 50 + "\n\n")
                f.write(current_deps)
            
            print(f"✓ Dependencies documented at: {dependencies_log_path}")
            return True
        else:
            print("✗ requirements.txt not found in project root")
            return False
    except Exception as e:
        print(f"✗ Error documenting dependencies: {str(e)}")
        return False

document_dependencies()
```
